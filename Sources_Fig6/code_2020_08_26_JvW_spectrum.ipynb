```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def Hamiltonian(a, b, eps, N):
    count=0
    
    hamiltonian = np.zeros((N,N))
    for x in range(0,N):
        for y in range(0,N):
            if count % 2 == 0:
                if y == x+1:
                    hamiltonian[x][y]=a
                if y == x-1:
                    hamiltonian[x][y]=-b
            else:
                if x == y+1:
                    hamiltonian[x][y]=a*(1-eps)
                if x == y-1:
                    hamiltonian[x][y]=-b*(1+eps)
        count=count+1
    
    return hamiltonian
```

```python
def eigen3(a,b,epsinit,epsfinal, steps, N):
    List_x=[]
    List_eps=[]
    List_vectors=[]
    List_eigval=[]
    List_lowest=[]
    temp2=[]
    
    epsmax=epsfinal-epsinit
    
    for i in range(0,N):
        List_x.append(i)
        
    for ii in range(0,steps+1):
        eps=epsinit+epsmax*ii/steps

        vectors=[]
        Ham = Hamiltonian(a, b, eps, N)
        eigen_val, eigen_vec = np.linalg.eig(Ham)
                
# Sort the eigenvalues by absolute values, along with the corresponding eigenvectors        
        for iii in range(0,N):
            for iiii in range(0,N-1):
                if abs(eigen_val[iiii]) > abs(eigen_val[iiii+1]):
                    temp = eigen_val[iiii+1]
                    eigen_val[iiii+1] = eigen_val[iiii]
                    eigen_val[iiii] = temp
                    
                    temp2[:] = eigen_vec[:,iiii+1]
                    eigen_vec[:,iiii+1] = eigen_vec[:,iiii]
                    eigen_vec[:,iiii] = temp2[:]
# end sort.

        List_eps.append(eps)
        List_eigval.append(abs(eigen_val))
        List_vectors.append(eigen_vec)

        
    return List_x, List_eps, List_eigval, List_vectors
```

parameters used before: b=1, a=1/2.5=0.4, eps=-0.425 (left and right localised zeromodes) or eps=-0.45 (two right localised zeromodes)

```python
x_coords, eps_coord, vals, vecs = eigen3(2.5,1.0,1.2,1.5,100,140)
```
